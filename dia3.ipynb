```python
#Challenge, crear una función que recibe dos números y suma

def suma_de_dos_numeros(numero1,numero2):
    numero1 = float(numero1)
    numero2 = float(numero2)
    print('La suma es', numero1+numero2)

suma_de_dos_numeros(4,8)
    
```

```python
#se puede definir dentro de los parámetros

def suma_de_dos_numeros2(float(numero11),float(numero22)):
    print(numero11+numero22)

suma_de_dos_numeros2(3,5)
```
