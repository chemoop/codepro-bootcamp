# Bucles infinitos
Para salir de un bucle infinito se usa *Break*

```python
numero3 = 30
while True:
    numero_a_ingresar = int(input("Ingresa un número"))
    if numero_a_ingresar == numero3:
        print("El número es correcto")
        break
    else:
        print("El número es incorrecto, inténtalo de vuelta")
```

```python
contraseña=input("Ingrese su contraseña")

while True:
    contraseña2 = input("Para ver el mensaje secreto, ingrese su contraseña")
    if contraseña == contraseña2:
        print("El mensaje secreto es amor y paz")
        break
    else:
        print("vuelva a introducir")
```

```python
lista_de_frutas = ["manzana","piña","pera"]
print(lista_de_frutas)

print(lista_de_frutas[0]) #Elegir elemento

lista_de_frutas.append("banana") #Para agregar
print(lista_de_frutas)

lista_de_frutas[2] = "cereza" #Modificar lista
print(lista_de_frutas)

lista_de_frutas.remove("manzana") #Eliminar elemento
print(lista_de_frutas)

lista_de_frutas.pop(0) #Eliminar por posición
print(lista_de_frutas)

print(len(lista_de_frutas)) #Cuenta los elementos de una lista
#También se puede definir una variable de la cantidad de elementos

zz = len(lista_de_frutas)
print(zz)
```
