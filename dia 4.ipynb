# Diccionarios
Es una colección de pares formados por una clave y un valor asociado a la clave. Se construyen poniendos los pares entre llaves {}

Características:
*No tienen orden
*Pueden contener elementos de distintos tipos
*Son mutables, pueden alterarse durante la ejecución de un programa
*Son únicas, no pueden repetirse en un mismo diccionario
*son iterables
*Los valores pueden ser de cualquier tipo

```python
mi_contacto = {'nombre': 'fulana','apellido': 'Britez', 'email' : 'dhfaf@gmail.com','telefono' : '098232435'}
print(mi_contacto)

mi_contacto ['nombre']

producto = {
    'nombre' : 'batería',
    'precio' : 450000, #integer
    'stock' : True #bool
}
#Acceder a una clave del diccionario
producto ['precio']
print(producto['precio'])

#modificar un elemento del diccionario
mi_contacto['email'] = 'penguin@gmail.com'
print(mi_contacto)

#usar metodos del diccionario.
print(mi_contacto.keys()) #devuelve todas las claves del diccionario
print(mi_contacto.values()) #devuelve todosd los valores del diccionario
print(mi_contacto.items())

#Añadir mas datos al diccionario
mi_contacto['ciudad'] : 'Asuncion'
print(mi_contacto)

#Editar datos
mi_contacto['nombre'] = 'Jose'
print(mi_contacto)


#Eliminar datos

#Eliminar el ultimo elemento (clave-valor) del diccionario
mi_contacto.popitem()
print(mi_contacto)

#Eliminar datos del diccionario por clave
del mi_contacto['apellido']
print(f'Este es el diccionario con la clave eliminada {mi_contacto}')

#Los diccionarios son iterables
for valores in mi_contacto.values():
    print(valores)
for claves in mi_contacto.keys():
    print(claves)
for clave,valor in mi_contacto.items():
    print(clave,valor)

#Otra manera de acceder a diccionarios es con dict()
d1 = {
  "Nombre": "Sara",
  "Edad": 27,
  "Documento": 1003882
}
print(d1)
```

# Programación orientada a objetos

En python se usa ***class*** el nombre de una clase siempre empieza en mayúscula

```python
#Molde
class Perro:
    # Atributos de clase
    cabeza = 1
    patas = 4

#creando un objeto -- Instanciar un objeto
firulais = Perro()
choco = Perro()

print(firulais.cabeza)
print(choco.patas)
```

```python
class Perro2:
    #Atributo de clase
    especie = 'canina'
    #Atributo de instancia
    def __init__(self, nombre, raza, color):
        self.nombre = nombre
        self.raza = raza
        self.color = color

#crear el objeto
firulais = Perro2('Firulais','mestizo','marron')
print(firulais.especie)

#Crear otro objeto
rompe = Perro2('Rompe','delmer','negro')
print(rompe.color)
print(rompe.especie)
```

```python
class Coche:
    tipo = 'vehiculo'
    #Atributos de instancia
    def __init__(self, modelo,posicion = 0):
        self.modelo = modelo
        self.posicion = posicion

    
    
    #Definir otras acciones de clase
    def acelerar(self,metros):
        self.posicion += metros
        print(f'El vehiculo modelo {self.modelo} acelera y avanza {metros} m. Nueva posición: {self.posicion} m')

    def girar(self, direccion):
        print(f'El {self.modelo} avanza y gira a la dirección {direccion}')

coche1 = Coche('Toyota')
coche1.acelerar(1)
coche1.girar('izquierda')
```

# Herencia

Dado que una clase hija hereda los atributos y metodos de la clase pafre, es util cuando tengamos clases que se parecen entre si pero tienen ciertas particularidades

```python
class AnimalDeGranja:
    #atributos de clase
    tipo = 'Animal de granja'

    #Atributos de instancia
    def __init__(self, nombre, color ,sonido):
        self.nombre = nombre
        self.color = color
        self.sonido = sonido
    #metodos (acciones de las clases)
    def hablar(self):
        print(f'Hablo y digo {self.sonido}')
    def presentarse(self):
        print(f'Hola, soy {self.nombre}')

class Gallina(AnimalDeGranja):
    huevos_por_dia = 1

    def poner_huevo(self):
        print(f'Estoy poniendo {self.huevos_por_dia} huevos por dia')

class Vaca(Gallina):
    patas = 4

    def mujir(self):
        print('Muuu')

gallina1 = Gallina('Hortencia','marron','corococo')
gallina1.hablar()
gallina1.presentarse()

vaca = Vaca('Lola','blanco y negro','Muuu')
vaca.hablar()
vaca.presentarse()
```

```python
class Libros:
    #atributo de clase
    tipo = 'Libro'
    
    #atributo de instancia
    def __init__(self, precio, tamaño): #init es para cosas que se pueden modificar, los de clase son fijos
        self.precio = precio
        self.tamaño = tamaño

#si se define dos veces init, python se queda con el ultimo
class CienciaFiccion(Libros):
        cantidad = 15
        def info(self):
            print(f'El precio del libro es {self.precio} y el tamaño del libro es {self.tamaño} y la cantidad en stock es {self.cantidad}')

libro1 = CienciaFiccion('50000', 'grande')

libro1.info()


```
